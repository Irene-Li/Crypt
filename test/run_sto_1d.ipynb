```python
import numpy as np
import sys, os 
sys.path.insert(1, os.path.join(sys.path[0], '..'))
from Source.CryptSto1D import *
import time
from matplotlib import pyplot as plt

# parameters of the differential equation
D1 = 0.25
D2 = 0.25
a = 1/D1 # want the fourth order derivative term to have coeff of 1 

phi0 = 0.1
n = 4

l = 4e-5
g = l
k = l
v0 = 10*l

epsilon = 1e-3

# simulation parameters
X = 128
n_batches = 100
phi_init = 0.5
f_init = 1
```

```python
solver = StoEvolution1D(epsilon, D1, D2, l, g, k, a, v0, phi0, n)
```

```python
T = 1e3
dt = 1e-2 
solver.initialise(X, T, dt, n_batches, phi_init, f_init)
start_time = time.time()
solver.evolve(verbose=True, cython=False)
end_time = time.time() 
print('time taken: ', end_time - start_time)
```

```python
solver.plot_phi_evol('test', show=True)
```

```python
plt.plot(solver.y[-2, solver.X:]) 
plt.title('f field')
plt.show() 
plt.plot(solver.y[-2, :X]) 
plt.title('phi field')
plt.show()
```
